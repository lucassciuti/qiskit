```python
import numpy as np
import matplotlib.pyplot as plt
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
from qiskit.circuit.library import MCMT
```

```python
n = 3  # 3 qubits -> lista com 8 elementos

open_circuit = QuantumCircuit(n)  # inicializando o circuito

# Colocando todos os qubits em superposição uniforme (Hadamart gates)
[open_circuit.h(qubit) for qubit in range(n)]
open_circuit.barrier()

# Oráculo para o item 110
"""open_circuit.x(2)
open_circuit = open_circuit.compose(MCMT('cz', 2, 1, label=None))
open_circuit.x(2)
open_circuit.barrier()"""

# Oráculo para o item 001
open_circuit.x(0)
open_circuit.x(1)
open_circuit = open_circuit.compose(MCMT('cz', 2, 1, label=None))
open_circuit.x(0)
open_circuit.x(1)
open_circuit.barrier()

# Diffuser
[open_circuit.h(qubit) for qubit in range(n)]
[open_circuit.x(qubit) for qubit in range(n)]

open_circuit.h(n - 1)
open_circuit.mct(list(range(n - 1)), n - 1)
open_circuit.h(n - 1)

[open_circuit.x(qubit) for qubit in range(n)]
[open_circuit.h(qubit) for qubit in range(n)]

# Fazendo a medida
open_circuit.measure_all()

open_circuit.draw(output='mpl')
```

```python
# usando um simulador de computador quântico do qiskit
backend = Aer.get_backend("qasm_simulator")

# executando o circuito 3000 vezes para gerar histograma
results = execute(open_circuit, backend, shots=3000).result()
answer = results.get_counts()

# plotando histograma
plot_histogram(answer)
```
